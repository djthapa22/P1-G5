```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
filepath = "Assessors_Parcel_Data_2022.csv"
home_value_data = pd.read_csv(filepath)
home_value_data.head()
```

```python

property_types = home_value_data[~home_value_data['PRIMARY_PROP_TYPE'].isin(['VACANT LAND - COMMERCIAL','COMMERCIAL','INDUSTRIAL',\
                                                         'VACANT LAND - INDUSTRIAL','MED/CARE FACILITY','COMMON AREA (NO VALUE)',\
                                                        'RES - ZERO LOT LINE'])]

new_property_types= property_types[property_types['TOTAL_UNITS'] > 0 ]
new_property_types= new_property_types[new_property_types['TOTALVALUE'] > 0 ]

# new_property_types= new_property_types.dropna()

```

```python
float_TV = new_property_types["TOTALVALUE"].astype('float64')
float_TU = new_property_types["TOTAL_UNITS"].astype('float64')

property_value_perunit= float_TV / float_TU

new_property_types["Per Unit Value"]= property_value_perunit

```

```python
new_property_types
```

```python
property_value = new_property_types.groupby('NEIGHBORHOOD').mean()['Per Unit Value']


property_value_df = pd.DataFrame({'Per Unit Value' : property_value})
property_value_df_sort = property_value_df.sort_values('Per Unit Value', ascending=False)
```

```python
top_5 = property_value_df_sort.head(5)
top_5
```

```python
bottom_5 = property_value_df_sort.tail(5)
bottom_5
```

```python
# Starting cleanup of Crime Data
file_2 = "crime_datacleaned.csv"
crime_data= pd.read_csv(file_2)
crime_data.head()
```

```python
crime_simple= crime_data[["Offense","NEIGHBORHOOD","Ward","Latitude","Longitude","Crime_Count","Reported_Date","Occurred_Date"]]
crime_simple.head()
```

```python
# Understanding general crime count and Crime %
offense_count= crime_simple.groupby("Offense")["Crime_Count"].sum()
total_count = crime_simple["Crime_Count"].sum()
offense_rate = (offense_count/ total_count)*100

total_crime= pd.DataFrame({"Crime Count by Offense":offense_count,"Total Crime %": offense_rate    
})
total_crime_sort = total_crime.sort_values("Total Crime %",ascending=False)
```

```python
total_crime_sort.head(10)
```

```python
total_crime_sort.tail(10)
```

```python
neighbor_count= crime_simple.groupby("NEIGHBORHOOD")["Crime_Count"].sum()
```

```python
neighbor_rate = (neighbor_count/ total_count) * 100
crime_metrics = pd.DataFrame({"NEIGHBORHOOD Crime Count":neighbor_count,"NEIGHBORHOOD Total Crime %":neighbor_rate
})
crime_metrics
```

```python
crime_metrics_sort= crime_metrics.sort_values("NEIGHBORHOOD Total Crime %",ascending=False)
```

```python
crime_metrics_sort.head(5)
```

```python
crime_metrics_sort.tail(5)
```
